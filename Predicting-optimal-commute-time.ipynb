```python
import pandas as pd

# Daniela was here...


# Read the Parquet file into a DataFrame
weather = pd.read_parquet('data1.parquet', engine='pyarrow')

# Ensure the date column is in datetime format, replace 'your_date_column' with the actual column name
weather['YYYYMMDD'] = pd.to_datetime(weather['YYYYMMDD'])

# Filter the DataFrame for dates from 2016 onwards
weather = weather[weather['YYYYMMDD'] >= '2018-01-01']

# Define the new column names
column_mapping = {
    'STN': 'Station',
    'YYYYMMDD': 'Date',
    'H': 'Hour',
    'DD': 'WindDirection',
    'FH': 'WindSpeedAvg',
    'FF': 'WindSpeed',
    'FX': 'WindGust',
    'T': 'Temperature',
    'T10N': 'MinTemp10cm',
    'TD': 'DewPoint',
    'SQ': 'SunshineDuration',
    'Q': 'GlobalRadiation',
    'DR': 'PrecipitationDuration',
    'RH': 'PrecipitationAmount',
    'P': 'AirPressure',
    'VV': 'Visibility',
    'N': 'CloudCover',
    'U': 'Humidity',
    'WW': 'Weather',
    'IX': 'WeatherCode',
    'M': 'Fog',
    'R': 'Rain',
    'S': 'Snow',
    'O': 'Thunder',
    'Y': 'IceFormation'
}

# Rename the columns
weather.rename(columns=column_mapping, inplace=True)

# Display the filtered DataFrame
weather
```
